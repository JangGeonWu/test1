# 그래프 스타일링
- 여러개의 그래프와 범례 표시하기

## 1 라이브러리 임포트 및 기본 설정

```python
import matplotlib.pyplot as plt
# 폰트 확인하기
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'Malgun' in f.name]
# '맑은 고딕'이 설치되어 있는지 검색
# 한글폰트 지정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 한글폰트 사용시 '-'기호가 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False # 원래 True로 되어있어 -기호가 깨짐
```

## 2. 여러개의 그래프 한번에 그리기
___
- 2명의 월별 몸무게 변화

```python
import pandas as pd
df1 = pd.DataFrame({'월':[1,2,3,4,5], '몸무게':[80,78,75,73,70]})
df2 = pd.DataFrame({'월':[1,2,3,4,5], '몸무게':[60,62,59,55,53]})
```

### 2.1. 여러 개의 그래프 각각 그리기
- plt.show()로 구분

```python
plt.plot(df1['월'], df1['몸무게'])
plt.show()

plt.plot(df2['월'], df2['몸무게'])
plt.show()
```

### 2.2. 서로 다른 그래프 한번에 그리기

```python
# 두 그래프 그리기
plt.plot(df1['월'], df1['몸무게'])
plt.plot(df2['월'], df2['몸무게'])

# 그래프 제목
plt.title('월별 몸무게 변화', fontsize=15)
# 축 라벨
plt.xlabel('월', loc='right')
plt.ylabel('몸무게(kg)', loc='top')
# x축 틱
plt.xticks(range(0, 14,1))
# y축 틱
plt.yticks(range(50, 91, 5))


plt.show()
```

## 2.3. 범례 표시하기
- 우선, 그래프에 레이블을 지정해주어야 한다(label=그래프명)
- plt.legend()

```python
# 두 그래프 그리기
plt.plot(df1['월'], df1['몸무게'], label='James')
plt.plot(df2['월'], df2['몸무게'], label='Amy')

# 그래프 제목
plt.title('월별 몸무게 변화', fontsize=15)
# 축 라벨
plt.xlabel('월', loc='right')
plt.ylabel('몸무게(kg)', loc='top')
# x축 틱
plt.xticks(range(0, 14,1))
# y축 틱
plt.yticks(range(50, 91, 5))
# 범례
plt.legend()

plt.show()
```

### 2.3.1. 범례 위치 지정
- plt.legend(loc=위치번호 or 위치이름)
___
|Location String|Location Code|
|--|--|
|'best'|0|
|'upper right'|1|
|'upper left'|2|
|'lower left'|3|
|'lower right'|4|
|'right'|5|
|'center left'|6|
|'center right'|7|
|'lower center'|8|
|'upper center'|9|
|'center'|10|

- tip: 0은 default, 1~4는 사분면을 의미
- plt.legend(loc=(x위치, y위치)) -> 왼쪽 하단이 기준점(0,0)

```python
# 두 그래프 그리기
plt.plot(df1['월'], df1['몸무게'], label='James')
plt.plot(df2['월'], df2['몸무게'], label='Amy')

# 그래프 제목
plt.title('월별 몸무게 변화', fontsize=15)
# 축 라벨
plt.xlabel('월', loc='right')
plt.ylabel('몸무게(kg)', loc='top')
# x축 틱
plt.xticks(range(0, 14,1))
# y축 틱
plt.yticks(range(50, 91, 5))
# 범례
# plt.legend(loc=4)
plt.legend(loc=(0.7, 0)) # 범위는 0~1 사이, 범위를 넘어서면 그래프 밖으로 뺄 수 있음

plt.show()
```

### 2.3.2 범례 속성 지정
- plt.legend(ncol=열개수)
- plt.legend(fontsize=폰트사이즈)
- plt.legend(frameon=True/False), 범례의 테투리 존재 유무
- plt.legend(shadow=True/False), 범례의 그림자 존재 유무
- plt.legend(facecolor = 색상)
- plt.legend(edgecolor = 색상)

```python
# 두 그래프 그리기
plt.plot(df1['월'], df1['몸무게'], label='James')
plt.plot(df2['월'], df2['몸무게'], label='Amy')

# 그래프 제목
plt.title('월별 몸무게 변화', fontsize=15)
# 축 라벨
plt.xlabel('월', loc='right')
plt.ylabel('몸무게(kg)', loc='top')
# x축 틱
plt.xticks(range(0, 14,1))
# y축 틱
plt.yticks(range(50, 91, 5))
# 범례
plt.legend(loc=4)
plt.legend(ncol=2, fontsize=12, frameon=True, facecolor='c', edgecolor='m', shadow=True)

plt.show()
```
