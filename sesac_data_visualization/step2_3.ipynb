## 2.1 y축 데이터로 그리기
- 1차원 리스트, 튜플, 시리즈 데이터를 전달하여 그린다
- x축은 데이터와 인덱스로 자동으로 지정된다

```python
# y축 데이터 [1,5,3,9,7]
data = [1,5,3,9,7]
plt.plot(data)
```

### 2.1.2 시리즈 데이터로 그리기

```python
import pandas as pd

# make series data
s = pd.Series([1,5,3,9,7])

s

# 왼쪽은 인덱스, 오른쪽은 값
```

```python
# graph
plt.plot(s)
```

```python
# 시리즈의 인덱스 변경하기
ss = pd.Series([1,5,3,9,7], index=[10,20,30,40,50])
ss
```

```python
# making graph
plt.plot(ss)
```

```python
# series의 인덱스 변경하기
sss = pd.Series([1,5,3,9,7], index=['a', 'b', 'c', 'd', 'e'])
sss
```

```python
# graph 그리기
plt.plot(sss)
```

## 2.2 x축, y축 데이터로 그리기
- x, y축의 데이터의 길이가 같아야 한다
---
### 2.2.1 리스트 데이터로 그리기

```python
x = [1,5,6,9,10]
y = [1,5,3,9,7]

plt.plot(x, y)
```

### 2.2.2 데이터프레임 데이터로 그리기

```python
ssss = pd.DataFrame({'x':[1,5,6,9,10], 'y':[1,5,3,9,7]})

ssss
```

```python
# graph
x = ssss['x']
y = ssss['y']

plt.plot(x, y)
```

### 2.2.3 시리즈 데이터로 그리기

```python
d = pd.Series([1,5,3,9,7])
d
```

```python
# series만 전달하여 그리기
plt.plot(d)
plt.show()
```

```python
# x, y 각각 전달하여 그리기
plt.plot(d.index, d.values)
plt.show() # text 보지 않고 그래프만 볼 수 있게 함
```
