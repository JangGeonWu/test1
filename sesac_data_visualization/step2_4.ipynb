# 2 시간의 시각화

```python
import matplotlib.pyplot as plt

# sample data: 어떤 아이의 나이별 키의 변화
age = [7,8,9,10,11,12,13,14,15]
height = [122.1, 127.9, 129.0, 130.1, 130.5, 133.2, 134.4, 137.7, 140.2]
```

# 3. 수량비교 시각화

```python
# 샘플 데이터: 100명중 혈액형 인원 비교
blood_type = ['a', 'b', 'o', 'ab']
count = [45,35,15,5]
```

```python
# 일반 막대 그래프
plt.bar(blood_type, count)
```

```python
# 가로 막대 그래프, barh
plt.barh(blood_type, count)
```

# 4. 비율의 시각화

```python
# 파이 차트
# plt.pie(count)
plt.pie(count, labels=blood_type)
plt.show()
```

# 5. 분포의 시각화, 히스토그램 ,바이올린 그래프 사용 

```python
# sample data: 1부터 100사이의 랜덤 점수 1000개
import numpy as np
data = np.random.randint(1, 101, 1000) # 범위 1 ~ 100
len(data)
```

```python
# 히스토그램
# plt.hist(data) # 지가 알아서 쪼갬
plt.hist(data, bins=100) # 100개로 쪼개서 보여줌
plt.show()
```

```python
# 상자 수염 그래프
plt.boxplot(data)
plt.show()
# 맨 아래는 최솟값, 맨 위는 최댓값, 중간의 막대(주황색 줄)는 중간값(50%), 막대의 아래부분은 25%, 윗부분은 75%를 의미
```

```python
# 바이올린 그래프
plt.violinplot(data)
plt.show()
```

# 6. 관계의 시각화

```python
# 샘플 데이터: 지불금액에 따른 팁
import seaborn as sns
# tips = sns.load_dataset('tips') # seaborn에서 제공하는 기본 데이터 셋
tips = sns.load_dataset('tips')[['total_bill', 'tip']] # 해당 데이터 셋 중 total_bill, tip 열만 가져옴
tips
```

```python
# 산점도
plt.scatter(tips['total_bill'], tips['tip'])
plt.show()
# 가로축 = total_bill(지불금액), 세로축=tip(팁)
```
