# 그래프 스타일링(제목, 축 레이블, 그리드)

```python
import matplotlib.pyplot as plt
# 폰트 확인하기
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'Gothic' in f.name]
# '맑은 고딕'이 설치되어 있는지 검색
```

```python
# 한글폰트 지정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 한글폰트 사용시 '-'기호가 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False # 원래 True로 되어있어 -기호가 깨짐
```

## 1.2 샘플 데이터
___
- 월별 몸무게 변화(1월: 80, 2월: 78, 3월: 75, 4월: 73, 5월: 70) 

```python
# 데이터 프레임 만들기
import pandas as pd
df = pd.DataFrame({'월':[1,2,3,4,5], '몸무게':[80,78,75,73,70]})
df
```

```python
# 2.1. 그래프 그리기
plt.plot(df['월'], df['몸무게'], 'ro-.', markerfacecolor='b', mec='b', lw=3, ms=7)
plt.show()
```

# 3. 제목
___
- plt.title('제목')
- loc(제목의 위치, location), left, right, center
- pad(타이틀과 그래프와의 간격, 완충제)
- color(폰트의 색상)
- fontsize(폰트 사이즈)
- fontweight(폰트 '굵기'), bold
- fontfamily(폰트, 예: 바탕체, 맑은 고딕, 나눔체 등등...)

```python
plt.plot(df['월'], df['몸무게'], 'ro-.', markerfacecolor='b', mec='b', lw=3, ms=7)
plt.title('월별 몸무게 변화', loc='left', pad = 10, color = 'r', fontsize=20, fontweight='bold', fontfamily='HYGothic-Extra')
plt.show()
```

# x축, y축 레이블
___
- plt.xlabel('레이블명')
- plt.ylabel('레이블명')
- loc(위치: left, center, right / bottom, center, top)
- labelpad: 레이블과 그래프와의 간격
- color: 폰트색상
- fontsize
- fontfamily

```python
# x, y축 레이블
plt.plot(df['월'], df['몸무게'], 'ro-.', markerfacecolor='b', mec='b', lw=3, ms=7)
plt.title('월별 몸무게 변화', loc='left', pad = 10, color = 'r', fontsize=20, fontweight='bold', fontfamily='HYGothic-Extra')


plt.xlabel('월', loc='right', color='r')
plt.ylabel('몸무게(kg)', loc='top', fontsize=20)
plt.show()
```

# 5. 그리드
___
## 5.1. 기본 그리드
- plt.grid(True)
- axis: 그리드 방향('x', 'y')
- ls, lw, color, alpha


```python
# x, y축 레이블
plt.plot(df['월'], df['몸무게'], 'ro-.', markerfacecolor='b', mec='b', lw=3, ms=7)
plt.title('월별 몸무게 변화', loc='left', pad = 10, color = 'r', fontsize=20, fontweight='bold', fontfamily='HYGothic-Extra')


plt.xlabel('월', loc='right', color='r')
plt.ylabel('몸무게(kg)', loc='top', fontsize=20)

plt.grid(True, axis='x')
plt.show()
```

```python
# x, y축 레이블
plt.plot(df['월'], df['몸무게'], 'ro-.', markerfacecolor='b', mec='b', lw=3, ms=7)
plt.title('월별 몸무게 변화', loc='left', pad = 10, color = 'r', fontsize=20, fontweight='bold', fontfamily='HYGothic-Extra')


plt.xlabel('월', loc='right', color='r')
plt.ylabel('몸무게(kg)', loc='top', fontsize=20)

plt.grid(True, color='c', alpha=0.5) # alpha= 투명도
plt.show()
```
