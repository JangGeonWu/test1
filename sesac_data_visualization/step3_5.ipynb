# 그래프 스타일링
___
## 한글폰트 사용

```python
import matplotlib.pyplot as plt
# 폰트 확인하기
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'Malgun' in f.name]
# '맑은 고딕'이 설치되어 있는지 검색
```

```python
# 한글폰트 지정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 한글폰트 사용시 '-'기호가 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False # 원래 True로 되어있어 -기호가 깨짐
```

```python
# 그래프로 확인
x = ['봄', '여름', '가을', '겨울']
y = [19, 30, -15, 20]

plt.plot(x, y)
plt.show()
```

## 3. 색상, 마커, 선
### 3.1. 색상(color)
___
#### 자주 사용되는 색깔: r, g, b, c(cyan), m(magenta), y, k(black), w(white)
- hex code를 사용할 수도 있다

```python
# 색상 이름
plt.plot(x, y, color='red')
plt.show()
```

```python
# 색상 코드
# 인터넷에서 '색상코드표'를 검색하여 색상 팔레트로부터 코드를 얻어냄

plt.plot(x, y, color = '#F361A6')
plt.show()
```

### 3.2. Marker
___
- 종류: (* . , o v ^ < > 1 2 3 4 s p * h H + x D d)
- markersize로 마커의 사이즈를 조절
- markeredgecolor(mec)로 마커 테두리의 색을 결정
- markeredgewidth(mew)로 마커 선의 굵기 결정
- markerfacecolor(mfc)로 마커 내부의 색을 결정

```python
plt.plot(x, y, marker='D', markersize='15', markeredgecolor='r', markeredgewidth='5', markerfacecolor = 'y')
plt.show()
```

## 3.3. 선
___
- linestyle(ls): 선 스타일, 종류='-', '--', '-.', ':'
- linewidth(lw): 선 굵기

```python
plt.plot(x, y, ls=':', lw='5')
plt.show()
```

## 3.4 [색상][마커][선]
___
- 일일이 위처럼 지정해주기는 귀찮으니까! [색][마커][선]이라는 포맷을 이용하자.

```python
plt.plot(x, y, 'k')
```

```python
plt.plot(x,y,'o')
```

```python
plt.plot(x, y, '--')
```

```python
# '색상:마커:선'으로 표현
plt.plot(x, y, 'ro:')
```

```python
plt.plot(x, y, 'ro:', markerfacecolor='k')
```
